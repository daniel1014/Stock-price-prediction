```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the main dataset about weather data from csv
df = pd.read_csv('MET Office Weather Data.csv',encoding='utf-8')

# filter the dataframe in the range of year 2000 to 2019
df = df[(df['year']>=2000) & (df['year']<=2019)]
df['year'] = df['year'].astype(int)
df['month'] = df['month'].astype(int)
df['dateInt']=df['year'].astype(str) + df['month'].astype(str) +'1'
df['date'] = pd.to_datetime(df['dateInt'], format='%Y%m%d')
df = df.drop(['dateInt'],axis=1)

df.head()
```

```python
# shows the general info of dataframe
df.info()
```

```python
# check duplicated value, no duplicated values is found
df[df.duplicated()]     
```

```python
# check if null value exist, i.e. 2532 null values have been found in attribute 'sun'
df.isnull().sum()
```

```python
# fill the missing values in numerical variables with the mean value of that column
df['sun'] = df['sun'].fillna(df['sun'].mean())
df['tmax'] = df['tmax'].fillna(df['tmax'].mean())
df['tmin'] = df['tmin'].fillna(df['tmin'].mean())
df['af'] = df['af'].fillna(df['af'].mean())
df['rain'] = df['rain'].fillna(df['rain'].mean())
```

```python
# Add an attribute mean temperature which is calculated from the average of the mean daily maximum and mean daily minimum temperature
df['tmean'] = (df['tmax']+df['tmin'])/2
```

```python
df.isnull().sum()   # We have successfully replaced all missing values
```

```python
# Explortary data analysis - mean temperature pattern according to stations
station_tmean = df.groupby('station')['tmean'].mean()
plt.rcParams['figure.figsize'] = 15,10
station_tmean.nlargest(36).plot(kind='bar')
```

```python
# using groupby to calculate the average value among all cities in monthly basis
df = df.groupby(['date']).agg({'year':'mean','month':'mean','tmax':'mean','tmin':'mean','tmean':'mean','af':'mean','rain':'mean','sun':'mean'}).reset_index()
df.head()
```

```python
# Reading from another Excel in order to extract values of total energy consumption in accordance to each month, which is our target attribute
with pd.ExcelFile('Electricity_ODS.xls') as xls:
    df_energy = pd.read_excel(xls, '5_3')
# Select appropraite columns and rows from the dataframe 
df_energy = df_energy.iloc[:,[0,1,2]]
df_energy = df_energy[4:-1]
# Rename column name
df_energy = df_energy.rename(columns={'5 ELECTRICITY':'year','Unnamed: 1':'month','Unnamed: 2':'energy_total'}).reset_index(drop=True)
# Filter the dataframe in the range of year 2000 to 2019
df_energy = df_energy[(df_energy['year']>=2000) & (df_energy['year']<=2019)].reset_index(drop=True)
df_energy
```

```python
# insert the target attribute energy_total into the main dataset in respect to each month from 2000 to 2019
df['energy_total'] = df_energy['energy_total']
df.head(10)
```

```python
df = df.astype({'energy_total':'float64'})
df.info()
```

```python
# check data staistic info, no abnormal value is found
df.describe().T.round(2)
```

```python
# Make bins and analysis relationship between amount of mean temperature and energy consumption 
bins_energy=[3,5,7,9]
bins_tmean=[0,6,12,18]
group=['Low','Average','High']
df['energy_bin']=pd.cut(df['energy_total'],bins_energy,labels=group)
df['tmean_bin']=pd.cut(df['tmean'],bins_tmean,labels=group)
energy_bin=pd.crosstab(df['tmean_bin'],df['energy_bin'])
energy_bin.div(energy_bin.sum(1).astype(float),axis=0).plot(kind='bar',stacked=True,figsize=(7,5))
plt.xlabel('Mean Temperature')
plt.ylabel('Proportion')
```

```python
# Make bins and analysis relationship between the amount of rainfall and energy consumption 
bins_rain=[0,60,120,190]
group=['Low','Average','High']
df['rain_bin']=pd.cut(df['rain'],bins_rain,labels=group)
rain_bin=pd.crosstab(df['rain_bin'],df['energy_bin'])
rain_bin.div(rain_bin.sum(1).astype(float),axis=0).plot(kind='bar',stacked=True,figsize=(7,5))
plt.xlabel('Rainfall')
plt.ylabel('Proportion')
```

```python
# drop the bins which we created for the exploration part
df = df.drop(['energy_bin','tmean_bin','rain_bin'], axis=1)
```

## Main part - Data Analytics Models

```python
# import library for cross-validation, logistic regression, and SVC
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.model_selection import TimeSeriesSplit
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
```

```python
# initialise blocking time series splitter 
class BlockingTimeSeriesSplit():
    def __init__(self, n_splits):
        self.n_splits = n_splits
    
    def get_n_splits(self, X, y, groups):
        return self.n_splits
    
    def split(self, X, y=None, groups=None):
        n_samples = len(X)
        k_fold_size = n_samples // self.n_splits
        indices = np.arange(n_samples)

        margin = 0
        for i in range(self.n_splits):
            start = i * k_fold_size
            stop = start + k_fold_size
            mid = int(0.5 * (stop - start)) + start
            yield indices[start: mid], indices[mid + margin: stop]
            
btscv = BlockingTimeSeriesSplit(n_splits=10)
```

```python
# initialise Time series splitter 
tscv = TimeSeriesSplit(n_splits=10)

# create and transform independent features into polynomial features
poly = PolynomialFeatures(degree=2,include_bias=False)    
poly_features = poly.fit_transform(X)

# split data into train and test data without shuffle 
x_train,x_test,y_train,y_test = train_test_split(X,y, test_size = 0.2, shuffle=False)

def regression(model,x,y):
    """ defining train and test data, train and evaluate corresponding regression models """
    x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, shuffle=False)
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)       # run model to predict y value
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))    
    print('RMSE is: ', rmse.round(3))    # the first performance measure - RMSE
    print('Accuracy is: ', model.score(x_test,y_test).round(3))  # the second performance measure - R2
    score = cross_val_score(model,x_train,y_train,cv=tscv,scoring='r2')   
    score2 = cross_val_score(model,x_train,y_train,cv=btscv,scoring='r2')               
    print('Time Series Cross validation Accuaracy: ', score.mean().round(3)) # the third performance measure - time series cross validation
    print('Blocking Time Series Cross validation Accuaracy: ', score2.mean().round(3))
    x = df['date']
    y = df['energy_total']
    plt.plot(x,y,color='blue',linewidth=3)      # plot actual value
    plt.plot(x[-48:],y_pred,color='red',linewidth=3)
    plt.legend(['observed','predicted'], fontsize=16)
```

```python
# create and test for Linear regression - poor result 
plt.rcParams['figure.figsize'] = 15,10
LG_model = LinearRegression()
regression(LG_model,X,y)  
```

```python
# create and test for Polynomial regression - good result
regression(LG_model,poly_features,y)
```

```python
# create and test for SVR   -- good result

SVR_model = SVR(kernel='linear')
regression(SVR_model,X,y)
```

```python
# test for another kernel on SVR  -- still shows good result when very low gamma was set
SVR_model = SVR(kernel='rbf', C= 100, gamma= 0.0001)
regression(SVR_model,X,y)
```
